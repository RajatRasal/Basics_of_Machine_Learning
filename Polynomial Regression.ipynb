# Polynomial Regression

```python
%matplotlib inline
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from random import randint

learning_rate = 0.01
training_epochs = 40
```

```python
data_points = 100
# 6 degree polynomial
poly_coefficients = [1, 2, 3, 4, 5, 6]
x_train = np.linspace(-1, 1, data_points)
y_train = 0
for i in range(len(poly_coefficients)):
    y_train += poly_coefficients[i] * np.power(x_train, i)
y_train += np.random.randn(len(y_train)) * 0.5
```

```python
plt.scatter(x_train, y_train)
plt.show()
```

```python
X = tf.placeholder("float")
Y = tf.placeholder("float")
```

```python
model = lambda X, w : tf.add_n([tf.multiply(w[i], tf.pow(X, i)) for i in range(len(poly_coefficients))])
```

```python
w = tf.Variable([0.] * len(poly_coefficients), name="parameters")
y_model = model(X, w)
```

```python
cost = tf.square(Y-y_model)
train_op = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```python
with tf.Session() as sess:
    init = tf.global_variables_initializer()
    sess.run(init)
    
    for epoch in range(training_epochs):
        for (x, y) in zip(x_train, y_train):
            print(x, y)
            sess.run(train_op, feed_dict={X: x, Y: y})
        
    w = sess.run(w)
    print(w)
```

```python
plt.scatter(x_train, y_train)
y_learned = 0
for i in range(len(poly_coefficients)):
    y_learned += w[i] * np.power(x_train, i)
plt.plot(x_train, y_learned, 'r')
plt.show()
```
